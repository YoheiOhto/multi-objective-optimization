## TDC dataset

```python
import pandas as pd
import math
from matplotlib import pyplot as plt
```

```python
path = "../../data/row/tdc/11_ClinTox.csv"
df = pd.read_csv(path)
```

```python
severity = df["Y"]
print(len(severity))
bins = math.ceil(math.log2(len(severity))) + 1
print(bins)
```

```python
title = path.replace("../../data/row/tdc/", "").replace(".csv", "")

plt.title(title)
plt.hist(severity, bins=bins)
```
